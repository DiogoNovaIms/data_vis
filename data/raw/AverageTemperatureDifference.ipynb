```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("GlobalLandTemperaturesByCountry+GHG.csv").drop(columns=["Unnamed: 0"])
df.head()
```

```python
na_countries = df.loc[df["continent"].isna()]["Country"].unique()
df.loc[df["Country"].isin(na_countries),"continent"] = "North America"
```

```python
df.loc[df["continent"].isna()]["Country"].unique()
```

```python
df = df.loc[df["year"]>=1900]
df.head()
```

```python
df_by_year = df.groupby(["year","Country","continent","country_code"]).mean().reset_index()
df_by_year
```

```python
def get_average_temperature_diff(country,year,temp):
    result = 0
    if year==1900:
        result = 0
    else:
        if(temp==np.nan):
            result = 0
        else:
            p = df_by_year.loc[(df_by_year["year"]==year-1) & (df_by_year["Country"]==country)]["AverageTemperature"]
            if(len(p)!=0):
                result= temp - p.tolist()[0]
    return result
        
df_by_year["temp_diff"] = df_by_year.apply(lambda x: get_average_temperature_diff(x.Country,x.year,x.AverageTemperature), axis=1)
```

```python
df_by_year
```

```python
df_by_year["continent"] = df_by_year["continent"].replace(["AS","SA","AF","OC","AN","EU","NA"],
                                                          ["Asia","South America","Africa","Oceania","Antartica","Europe",
                                                          "North America"])
df_by_year
```

```python
def get_temperature_diff_between(df_by_year,start_date,end_date):
    df = df_by_year.loc[(df_by_year["year"] >= start_date) & 
                        (df_by_year["year"] <= end_date)].groupby(["Country"]).sum()[["temp_diff"]]
    
    return df
    
```

```python
get_temperature_diff_between(df_by_year,1900,2000)
```

```python
df_by_year.to_csv("CountryTemperatures_byYear.csv")
```
