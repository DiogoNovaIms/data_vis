```python
import pandas as pd
import numpy as np
from pycountry_convert import country_alpha2_to_continent_code, country_name_to_country_alpha2
```

```python
df_1900 = pd.read_csv("1900_2021_DISASTERS.xlsx - emdat data.csv")
df_1900
```

```python
df_1900.columns
```

```python
df = df_1900[["Year","Event Name","Disaster Type","Disaster Group","Disaster Subgroup","Country",
              "Continent","Total Deaths","Total Affected","End Day","End Year","End Month"]]

df = df[~df['End Day'].isna()]
df["End Day"] = df["End Day"].astype(int)
df["End Month"] = df["End Month"].astype(int)
df["End Year"] = df["End Year"].astype(int)

df["Date"] = df["End Year"].astype(str) + "-" + df['End Month'].astype(str) + "-" + df['End Day'].astype(str) 
df["Date"] = pd.to_datetime(df["Date"],errors="coerce")
df = df[~df['Date'].isna()]

df
```

```python
df["Country"] = df["Country"].str.replace("\(the\)","")
df["Country"] = df["Country"].str.rstrip()
```

```python
def get_country_code(x):
    result = np.nan
    try:
        result = country_name_to_country_alpha2(x,cn_name_format="upper")  
    except:
        result = np.nan
    return result
    
def get_continent(x):
    result = np.nan
    try:
        result = country_alpha2_to_continent_code(x)
    except:
        result = np.nan
    return result
```

```python
df = df[~df['Country'].isin(list_eliminate+["Timor-Leste"])]

df = df.replace(
   ['Iran (Islamic Republic of)','Taiwan (Province of China)','Congo (the Democratic Republic of the)',
    'Korea (the Republic of)','Bolivia (Plurinational State of)',
    'United Kingdom of Great Britain and Northern Ireland (the)',
   'Venezuela (Bolivarian Republic of)','Macedonia (the former Yugoslav Republic of)','Moldova (the Republic of)',
   "Korea (the Democratic People's Republic of)",'Côte d’Ivoire'],
   ['Iran','Taiwan','Democratic Republic of the Congo',
   'South Korea','Bolivia','United Kingdom','Venezuela','North Macedonia','Moldova',"North Korea",'Ivory Coast'])

df.head()
```

```python
df["country_code"] = df["Country"].apply(lambda x: get_country_code(x))
df["continent"] = df["country_code"].apply(lambda x: get_continent(x))
```

```python
missing = df[df['country_code'].isna()]
list_eliminate = missing["Country"].value_counts().index.tolist()
list_eliminate
```
