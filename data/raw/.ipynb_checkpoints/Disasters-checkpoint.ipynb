```python
df = df_1900[["Year","Disaster Type","Disaster Group","Disaster Subgroup","Country","Continent","Total Deaths"]]
df
```

```python
df["Country"] = df["Country"].str.replace("\(the\)","")
df["Country"] = df["Country"].str.rstrip()
```

```python
def get_country_code(x):
    result = np.nan
    try:
        result = country_name_to_country_alpha2(x,cn_name_format="upper")  
    except:
        result = np.nan
    return result
    
def get_continent(x):
    result = np.nan
    try:
        result = country_alpha2_to_continent_code(x)
    except:
        result = np.nan
    return result
```
