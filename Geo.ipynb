```python
import pandas as pd
import json
```

```python
original_df = pd.read_csv("GlobalLandTemperaturesByCountry+GHG.csv").drop("Unnamed: 0",axis=1)
original_df.head()
```

```python
df = original_df.loc[original_df["dt"]=="2008-01-01"]
df
```

```python
with open("custom.geo.json",encoding='utf-8') as f:
    geo_world = json.load(f)
```

```python
found = []
missing = []
countries_geo = []

tmp = df.set_index("Country")

for country in geo_world["features"]:
    
    #Detect the country code
    country_code = country["properties"]["name_long"].upper()
    
    if(country_code in tmp.index):
        
        found.append(country_code)
        
        geometry = country["geometry"]
        
        countries_geo.append({
            "type": "Feature",
            "geometry": geometry,
            "id": country_code
        })
        
        
    else:
        
        missing.append(country_code)
        
print(f'Countries found    : {len(found)}')
print(f'Countries not found: {len(missing)}')
geo_world_ok = {'type': 'FeatureCollection', 'features': countries_geo}
```
