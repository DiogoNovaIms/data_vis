```python
import pandas as pd
```

```python
emissions = pd.read_csv("emission data.csv")
emissions.head()
```

```python
#Melt the columns into rows
emissions_df = pd.melt(emissions, id_vars=["Country"], 
                  var_name="year", value_name="GHG")

emissions_df["Country"] = emissions_df["Country"].str.upper()
emissions_df["year"] = emissions_df["year"].astype(int)

emissions_df.head()
```

```python
temp_df = pd.read_csv("GlobalLandTemperaturesByCountry_clean.csv").drop("Unnamed: 0", axis = 1)

temp_df["Country"] = temp_df["Country"].astype(str)
temp_df["year"] = temp_df["year"].astype(int)

temp_df.head()
```

### Merge the two datasets by country and year

```python
df = pd.merge(temp_df,emissions_df,on=["Country","year"],how="left")
df.tail(50)
```

### Export data

```python
df.to_csv("GlobalLandTemperaturesByCountry+GHG")
```
